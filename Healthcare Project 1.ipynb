```python
# Finding the Maximum And Minimun Age of the dataset

min_age = data['Age'].min()
max_age = data['Age'].max()

print(f"The minimum age in the dataset is: {min_age} years")
print(f"The maximum age in the dataset is: {max_age} years")
```

```python
#Creating the function to finding Diseases for Random Age Group

def diseases_by_age_gender(data):
    age_min = int(input("Enter the minimum age: "))
    age_max = int(input("Enter the maximum age: "))
   
    filtered_data = data[(data['Age'] >= age_min) & (data['Age'] <= age_max)]

    if filtered_data.empty:
        print("No data available for the specified age range.")
        return
    
    grouped_data = filtered_data.groupby(['Gender', 'Medical Condition']).size().unstack(fill_value=0)

    print(f"Distribution of Diseases for Age Group {age_min} to {age_max}:")
    print(grouped_data)
    

    sns.set(style="whitegrid")
    plt.figure(figsize=(12, 6))
# Plotting the graph
    sns.countplot(data=data, x='Medical Condition', hue='Gender', palette='Set1')
    plt.title("Distribution of Medical Conditions by Gender")
    plt.xlabel("Medical Condition")
    plt.ylabel("Number of Patients")
    plt.xticks(rotation=90)
    plt.legend(title='Gender')
    plt.show()
    
    
    # Plotting using seaborn Bar Plot
    plt.figure(figsize=(10, 6))
    sns.barplot(data=grouped_data, palette="YlGnBu")
    plt.title(f"Distribution of Diseases for Age Group {age_min} to {age_max}")
    plt.xlabel("Medical Condition")
    plt.ylabel("Count")
    plt.show()
    

diseases_by_age_gender(data)
```

```python

# making Age group 10-10 

data['Date of Admission'] = pd.to_datetime(data['Date of Admission'])

filtered_data = data[(data['Date of Admission'].dt.year >= 2018) & (data['Date of Admission'].dt.year <= 2023)]

age_bins = [18, 27, 37, 47, 57, 67, 77, 85]
age_labels = ['18-27', '28-37', '38-47', '48-57', '58-67', '68-77', '78-85']

filtered_data['Age Group'] = pd.cut(filtered_data['Age'], bins=age_bins, labels=age_labels, right=False)


grouped_data = filtered_data.groupby(['Age Group', 'Medical Condition']).size().unstack(fill_value=0)


print("Distribution of Diseases by Age Group from 2018 to 2023:")
print(grouped_data)


# Plot the distribution of diseases among different age groups
plt.figure(figsize=(12, 8))
grouped_data.plot(kind='bar', stacked=True)
plt.title("Distribution of Diseases by Age Group from 2018 to 2023")
plt.xlabel("Age Group")
plt.ylabel("Number of Cases")
plt.xticks(rotation=45, ha='right')
plt.legend(title='Medical Condition', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()

```

```python
# making a function to give input to pirticular input and getting the output gender(male,female) and their diseases

def medical_condition_gender_by_blood_type(data):
    blood_type_input = input("Enter the blood type (O-, O+, B-, AB+, A+): ").upper()
    
  
    filtered_data = data[data['Blood Type'] == blood_type_input]

    if filtered_data.empty:
        print("No data available for the specified blood type.")
        return
    

    grouped_data = filtered_data.groupby(['Medical Condition', 'Gender']).size()

    print(f"Medical Condition and Gender Distribution for Blood Type {blood_type_input}:")
    print(grouped_data)

    
    
    
# Plotting the graph
    plt.figure(figsize=(12, 6))
    sns.countplot(data=filtered_data, x='Medical Condition', hue='Gender', palette='Set1')
    plt.title(f"Medical Condition Distribution by Gender for Blood Type {blood_type_input}")
    plt.xlabel("Medical Condition")
    plt.ylabel("Number of Patients")
    plt.xticks(rotation=45, ha='right')
    plt.legend(title='Gender')
    plt.show()
```

```python
# the function on giving the input of two age and get the output on this age group whivh diseases are count the blood group

def medical_condition_blood_type_relationship(data):
    age1 = int(input("Enter the first age: "))
    age2 = int(input("Enter the second age: "))
    
    # Filter the data for patients within the specified age range
    filtered_data = data[(data['Age'] >= age1) & (data['Age'] <= age2)]
    
    if filtered_data.empty:
        print("No data available for the specified age range.")
        return
    
    # Group data by 'Medical Condition' and 'Blood Type', then calculate the count of patients
    grouped_data = filtered_data.groupby(['Medical Condition', 'Blood Type']).size().unstack(fill_value=0)

    print(f"Relationship between Medical Condition and Blood Type for patients aged {age1} to {age2}:")
    print(grouped_data)
    
    # Plotting the graph
    plt.figure(figsize=(12, 6))
    sns.heatmap(grouped_data, cmap='YlGnBu', annot=True, fmt='g', linewidths=0.5)
    plt.title(f"Relationship between Medical Condition and Blood Type for patients aged {age1} to {age2}")
    plt.xlabel("Blood Type")
    plt.ylabel("Medical Condition")
    plt.show()
```

```python
medical_condition_blood_type_relationship(data)
```
